```python
import pandas as pd

df = pd.read_excel('4907-Sample-02.xlsx')
print(df)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (18,9), facecolor='lightblue')
plt.xlabel('bbb')
plt.ylabel('yyyyys')
plt.plot('X','Y1' ,'d-b'  , data=df)
plt.plot('X','Y2' ,'d-g'  , data=df)
plt.legend()
plt.show()
fig.savefig('fig1.png')
```

```python
### matplotlibでグラフを保存
import numpy as np
import matplotlib.pyplot as plt
import pathlib

x1 = np.random.normal(0, 10, 1000)
x2 = np.random.normal(20, 20, 1000)

fig = plt.figure(figsize = (8,5), facecolor='lightblue')

plt.xlabel('X')
plt.ylabel('Y')
plt.hist(x1,  bins=30, range=(-50, 100), ec='black', alpha = 0.5, label='x1')  #binsは本数
plt.hist(x2,  bins=30, range=(-50, 100), ec='black', alpha = 0.5, label='x2')
plt.legend()
plt.show()

# path_dir = pathlib.Path(r'test')
# path_img = path_dir.joinpath('img.png')
# fig.savefig(path_img)

fig.savefig('ty.png')
```
